```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
df = pd.read_csv('car_info_new.csv')
```

```python
sns.set_theme(style="darkgrid")
g = sns.jointplot(x="Mean", y='Uniques', data=df, xlim=12000, ylim=30, palette="gist_ncar_r", kind = "hex")
```

```python
sns.set_theme(style="darkgrid")
g = sns.histplot(df, binwidth=100, kde=True)
g.set(title='Histplot')
g.set(xlim=(0, 12000))
g.set(ylim=(0, 50))
```

```python
sns.set_theme(style="darkgrid")
g = sns.lineplot(data=df)
g.set(title='Lineplot')
g.set(xlim=(0, 50))
g.set(ylim=(0, 16000))
```

```python
df.corr()
```

```python
sns.set_theme(style="darkgrid")
g = sns.pairplot(df)
g.set(xlim=(0, 100))
g.set(ylim=(0, 16000))
g.map_upper(sns.scatterplot, s=15)
g.map_diag(sns.kdeplot, lw=2)

```

```python
sns.set_theme(style="ticks")
dots = sns.load_dataset("dots")
g = sns.relplot( 
    data=df,
    x="Uniques", y="Mean",
    hue="Uniques",
    kind="line", palette="gist_ncar_r",
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
g.set(xlim=(0, 50))
g.set(title='Relplot')
```
